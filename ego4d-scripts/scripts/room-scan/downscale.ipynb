```python
import cv2
import numpy as np
from pathlib import Path
```

```python
original_folder = Path("images/")
resized_folder = Path("images_lowres")
```

```python
num_imgs = 855
for id in range(1, num_imgs+1):
    file_name = "{:06d}.png".format(id)
    print(file_name)
    original_img_name = str(original_folder / file_name)

    original_img = cv2.imread(original_img_name)

    #########################################################

    scale_percent = 10 # percent of original size
    width = int(original_img.shape[1] * scale_percent / 100)
    height = int(original_img.shape[0] * scale_percent / 100)
    dim = (width, height)
    
    resized_img = cv2.resize(original_img, dim, interpolation = cv2.INTER_AREA)
    resized_img_name = str(resized_folder / file_name)

    cv2.imwrite(resized_img_name, resized_img)
    print("Saved image: " + resized_img_name)

```
