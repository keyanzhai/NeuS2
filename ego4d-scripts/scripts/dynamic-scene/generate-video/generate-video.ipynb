```python
# Use Kernel "human_pose"
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import json
from pathlib import Path
import os
```

```python
source_img_path = Path("/home/kz/Desktop/Research/NeuS2-official/NeuS2/data/ego4d/dynamic-scene/images")
```

```python
target_video_path = Path("/home/kz/Desktop/Research/NeuS2-official/data-preprocess/Ego4d/raw-data/upenn_0715_Piano_1_4/dynaimc_scene/videos")
```

```python
# camera_list = ["aria01_rgb", "aria01_slam-left" , "aria01_slam-right", "gp01_0", "gp02_0", "gp03_0", "gp04_0", "gp05_0"]
camera_list = ["gp01_0", "gp02_0", "gp03_0", "gp04_0", "gp05_0"]
```

```python
# For each camera from ["gp01_0", "gp02_0", "gp03_0", "gp04_0", "gp05_0"]
for camera_name in camera_list:
    # The output video for the current camera
    out_video_name = camera_name + ".mp4"
    out_video_path = str(target_video_path / out_video_name)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    # First frame (12960)
    # get the dimensions of the first image
    frame_idx = 12960
    frame_name = "{:06d}".format(frame_idx)
    img_name = frame_name + "_" + camera_name + ".jpg"
    img_path = str(source_img_path / frame_name / img_name)
    img = cv2.imread(img_path)
    height, width, channels = img.shape

    # create the VideoWriter object
    out = cv2.VideoWriter(out_video_path, fourcc, 10, (width, height))

    # Read all the remaining frames [12961, 13109]
    for frame_idx in range(12961, 13109 + 1): 
        frame_name = "{:06d}".format(frame_idx)
        img_name = frame_name + "_" + camera_name + ".jpg"
        img_path = str(source_img_path / frame_name / img_name)
        img = cv2.imread(img_path)
        out.write(img)

        print("Successfully written to video: ", img_name)

    out.release()

    print("Successfully written video: ", out_video_name)

```
